```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv('../data/cleaned.csv')
```

```python
df['model'].value_counts()
```

```python
prices = df.groupby('model').price.mean()
```

```python
high_price = prices.sort_values().tail(10).plot(kind='bar', title='Highest Mean Price per Model', figsize=(8, 10), ylim=(0, 90000));
high_price.figure.savefig('../images/high_price.png', bbox_inches='tight')
```

```python
low_price = prices.sort_values().head(10).plot(kind='bar', title='Lowest Mean Price per Model', figsize=(8, 10), ylim=(0, 90000));
low_price.figure.savefig('../images/low_price.png', bbox_inches='tight')
```

```python
(df.groupby('model').price.max() - df.groupby('model').price.min()).mean()
```

```python
range = df.groupby('model').price.max() - df.groupby('model').price.min()
```

```python
ranges = range.sort_values().tail(10).plot(kind='bar', title='Highest Price Ranges per Model (difference between the cheapest and most expensive car)', figsize=(8, 10));
ranges.figure.savefig('../images/range.png', bbox_inches='tight')
```
